```python
# How do you calculate the total sum of 'Salary' for each 'Department' in a dataset?
```

```python
df.groupby('Department')['Salary'].sum()
```

```python
## How can you count the number of employees in each 'Department' in a dataset?
```

```python
df.groupby('Department')['Name'].count()
```

```python
## How do you compute the average 'Age' for each 'Gender' in a dataset?
```

```python
df.groupby('Gender')['Age'].mean()
```

```python
## Calculate both the total and average 'Salary' for each 'Gender'?
```

```python
df.groupby('Gender')['Salary'].agg(['sum', 'mean'])
```

```python
## How do you calculate the total 'Salary' for each combination of 'Department' and 'Gender' in a dataset?
```

```python
df.groupby(['Department', 'Gender'])['Salary'].sum()
```

```python
## How do you filter out 'Departments' where the average 'Salary' is below 60,000 in a dataset?
```

```python
df.groupby('Department')['Salary'].mean() < 60000
```

```python
## How do you sort the DataFrame by 'Age' in ascending order?
```

```python
df.sort_values(by='Age')
```

```python
## How do you sort the DataFrame first by 'Department' in ascending order, and then by 'Salary' in descending order?
```

```python
df.sort_values(by=['Department', 'Salary'], ascending=[True, False])
```

```python
## How do you sort the DataFrame by 'JoiningDate' in ascending order, ensuring NaN values are placed at the end?
```

```python
df.sort_values(by='Age', ascending=False).reset_index(drop=True)
```

```python
## How can you add a new column 'Bonus' that is 5% of the 'Salary' for each employee in a DataFrame?
```

```python
df['salary_bonus'] = df['Salary'] * 0.05
```

```python
df
```

```python
## How do you rename the columns ['ID', 'JoiningYear'] to ['EmployeeID', 'YearJoined'] in a DataFrame?
```

```python
df.rename(columns={'ID': 'EmployeeID', 'JoiningYear': 'YearJoined'}, inplace=True)
```

```python
df
```

```python
## How can you add a new column 'Seniority' that labels employees as 'Senior' if their 'Age' is above 40, and 'Junior' otherwise, in a DataFrame?
```

```python
df['Seniority'] = df['Age'].apply(lambda x: 'Senior' if x > 40 else 'Junior')
```

```python
df
```

```python
## How do you rename all columns to lowercase in a DataFrame?
```

```python
df.rename(columns=str.lower, inplace=True)
```

```python
df
```

```python
## How can you replace all occurrences of 'Male' with 'M' in the 'Gender' column of a DataFrame?
```

```python
df.replace({'Gender': {'Male': 'M'}}, inplace=True)
```
